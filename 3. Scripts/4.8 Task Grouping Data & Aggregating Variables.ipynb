```python
#3 steps combined to analyse customer loyalty by number of orders
df_ords_prods_merged['max_order'] = df_ords_prods_merged.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
#Check output
df_ords_prods_merged.head()
```

```python
#Create loyalty flag criteria - loyal customer
df_ords_prods_merged.loc[df_ords_prods_merged['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
#Create loyalty flag criteria - regular customer
df_ords_prods_merged.loc[(df_ords_prods_merged['max_order'] <= 40) & (df_ords_prods_merged['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
#Create loyalty flag criteria - new customer
df_ords_prods_merged.loc[df_ords_prods_merged['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
#Loyalty flag groupings count
df_ords_prods_merged['loyalty_flag'].value_counts(dropna = False)
```

```python
#Miultiple aggregations - average / min / max product prices by customer loyalty group
df_ords_prods_merged.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']})
```

```python
#3 steps combined to analyse customer spending by average cost per order
df_ords_prods_merged['avg_order_cost_by_cust'] = df_ords_prods_merged.groupby(['user_id'])['prices'].transform('mean')
```

```python
df_ords_prods_merged.head()
```

```python
#Create spender flag criteria - low spender
df_ords_prods_merged.loc[df_ords_prods_merged['avg_order_cost_by_cust'] < 10, 'spender_flag'] = 'Low spender'
```

```python
#Create spender flag criteria - high spender
df_ords_prods_merged.loc[df_ords_prods_merged['avg_order_cost_by_cust'] >= 10, 'spender_flag'] = 'High spender'
```

```python
df_ords_prods_merged.head(1500)
```

```python
#3 steps combined to analyse customer spending by average cost per order
df_ords_prods_merged['cust_frequency'] = df_ords_prods_merged.groupby(['user_id'])['days_since_prior_order'].transform('median')
```

```python
df_ords_prods_merged.head()
```

```python
#Create customer frquency flag criteria - non-frequent
df_ords_prods_merged.loc[df_ords_prods_merged['cust_frequency'] > 20, 'frequency_flag'] = 'Non-frequent customer'
```

```python
#Create customer frequency flag criteria - regular customer
df_ords_prods_merged.loc[(df_ords_prods_merged['cust_frequency'] <= 20) & (df_ords_prods_merged['cust_frequency'] > 10), 'frequency_flag'] = 'Regular customer'
```

```python
#Create customer frquency criteria - new customer
df_ords_prods_merged.loc[df_ords_prods_merged['cust_frequency'] <= 10, 'frequency_flag'] = 'Frequent customer'
```
